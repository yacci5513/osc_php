### 집합
- 리스트와 같이 여러개의 데이터를 집어넣수 있으나, **중복이 불가**

- 데이터 내 순서 X
- 집합연산(교집합, 합집합)이 가능

````python
set1 = {1,2,3}
set2 = {2,3,4}
print(set1 & set2) # 교집합(and)
print(set1 | set2) # 합집합(or)
print(set1 - set2) # 차집합
print(set2 - set1) # 차집합
````

# 제어문(조건문)
- python에서는 조건문을 if물 활용한다.
  - 조건에 종속되는 문장은 들여쓰기를 이용해 종속관계를 표현한다.
  ```
  if 조건문:
      종속문장
    ```
- if-else를 이용해 if 조건이 해당되지 않을 때 시행 될 종속문장을 실행할 수 있다.
  ```
  if 조건문:
      종속문장
  else:
      종속문장
  ```
- 여러 조건을 주고 싶을 때는 elif를 활용하면 된다.
  ```
  if 조건문1:
      종속문장
  elif 조건문2:
      종속문장
  else:
      종속문장
  ```
- elif는 개수제한없이 사용가능하다.
- 조건문은 비교연산자와 함께 활용된다.
  ```
  x = 3
  y = 2

  x > y # True
  x < y # False
  x >= y # True
  x <= y # False
  x == y # False
  X != y # True
  ```
- 만약 하나의 조건문에 여러 조건을 주고 싶은 경우에는 논리연산자를 활용
  ```
  x = True
  y = False

  x or y : x와 y 둘중에 하나만 참
  x and y : x와 y 둘다 참
  not x : x가 거짓이면 참이다.

  x and y # False
  x or y # True
- list에 element가 포함 되어 있는지를 in과 함께 확인할 수 있다.
  ```

````python
# 돈이 3000원 있거나 카드가 있다면 택시를 타고 그렇지 않으면 걸어가라
money = 2000
card = 1

if (money >= 3000) or (card):
  print('택시를 타라')
else:
  print('걸어 가라')
````

# 반복문
- 특정 기능을 반복 수행할 때 사용
- 특정 조건이 맞으면 계속 반복
- 들여쓰기를 이용해 반복 조건에 대한 수행 할 종속문을 표현
```
while 조건문:
    종속문
```
- 파이썬에서 반복문은 while, for문이 있다.

## while 문
  ```
  while 조건:
      수행할 문장 1
      수행할 문장 2
  ```
- while문은 조건에 해당되는 경우 계속 실행하기 때문에 break와 함께 사용
- while 문에 포함 된 조건문을 사용하다가 break를 만나면 반복문이 종료

````python
 # 열 번 찍어 안 넘어 가는 나무 없다.
treeHit = 0
while treeHit < 10:
  treeHit = treeHit + 1
  print(f"나무를 {treeHit}번 찍었습니다.")
  if treeHit == 10:
    print('나무가 넘어갑니다.')
````

````python
# break 활용하기
coffee = 10
money = 300

while money:
  print("돈을 받았으니 커피를 줍니다.")
  coffee = coffee - 1
  print(f"남은 커피의 양은 {coffee}개 입니다.")

  if not coffee:
    print('커피가 다 떨어졌습니다.')
    break
````

## for 문

````python
list1 = ['one', 'two', 'three']
for i in list1:
  print(i)
````

````python
list2 = [(1,2), (3,4), (5,6)]
for (num1, num2) in list2:
  print(num1, num2)
````

````python
# for
marks = [90, 25, 67, 45, 80]

student_num = 0
for mark in marks:
  student_num = student_num + 1
  if mark >= 60:
    print(f'{student_num}번 학생은 합격입니다.')
  else:
    print(f'{student_num}번 학생은 불합격입니다.')
````

````python
# for + continue(현재 반복을 중지하고 다음 반복으로 넘어가는 역할)
marks = [90, 25, 67, 45, 80]

student_num = 0
for mark in marks:
  student_num = student_num + 1 # 4, 45
  if mark < 60:
    continue
  print(f'{student_num}번 학생은 합격입니다.')
````

````python
# 이중포문
list1 = [1,2,3]
list2 = [4,5,6]

for num1 in list1:
  for num2 in list2:
    print(num1, num2)
````

````python
# 리스트 컴프리션(for)
a = [1,2,3,4]
result = [i*3 for i in a]
result
````

````python
# 리스트 컴프리션(for + if)
result2 = [i*3 for i in a if i % 2 == 0] # 2, 4
result2
````

# 함수

```

    <함수를 호출 했을 때 수행할 문장>
    return 출력 Y
```

Example 2-10. 두 수를 입력 받아 사칙연산(+,-, x, /)을 하는 함수 만들기

````python
def cal(num1, num2):
  add = num1 + num2
  sub = num1 - num2
  mul = num1 * num2
  div = num1 / num2
  return add, sub, mul, div
cal(5, 10)
````
