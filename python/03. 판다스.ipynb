### 데이터 선택(***)
- loc  : index, column으로 데이터를 가져온다.
- iloc : 데이터의 순서로 가져온다.

```
df2.loc[row의 이름, column의 이름]
df2.iloc[row의 순서, column의 순서]
```

````python
print(df2.iloc[0, 0]) # 44000
print(df2.iloc[1, 2]) # 85500
print(df2.iloc[2, 1:])# 70300,86000
print(df2.iloc[1:, 1]) # 70200, 70300
````

````python
df2.iloc[1] # = df2.iloc[1]
df2.iloc[:, 1]
````

````python
print(df2.loc['2023-01-01', 'SK텔레콤'])
print(df2.loc['2023-01-02', 'LG전자'])
print(df2.loc['2023-01-03', '삼성전자':])
print(df2.loc['2023-01-02':, '삼성전자'])
````

### 데이터프레임 결합
- concat : 그냥 가져다 붙이는 거
- merge :  공통된 칼럼이나 인덱스가 있는 경우

#### conat

````python
dates2 = pd.date_range('20230104', periods = 3)
prices = {
    'SK텔레콤' : [44100, 45500, 46000],
    '삼성전자' : [70200, 71200, 71300],
    'LG전자' : [85100, 85600, 84000],
    'POSCO' : [451000, 456000, 450000]
}

df3 = pd.DataFrame(prices, index = dates2)
df3
````

````python
concat_df1 = pd.concat([df2, df3])
concat_df1
````

````python
dates3 = pd.date_range('20230101', periods = 6)
s1 = pd.Series([500000, 501000, 505000, 500000, 501000, 505000], index = dates3, name = 'LG화학')
s1
````

````python
concat_df2 = pd.concat([concat_df1, s1], axis = 1)
concat_df2
````

#### merge
- inner : 양쪽 데이터프레임에 on으로 작성해준 공통 칼럼(key)가 있는 경우

````python
market_price_change = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/market_price_change_2022.csv')
purchases_of_equities = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/purchases_of_equities_2022.csv')
print(market_price_change.shape)
print(purchases_of_equities.shape)
````

````python
market_price_change.sort_values('티커')
````

````python
purchases_of_equities.sort_values('티커')
````

````python
merge1 = pd.merge(left = market_price_change, right =purchases_of_equities,
                  on = ['티커', '종목명']) # inner
print(merge1.shape)
merge1.head()
````

````python
merge3 = pd.merge(left = market_price_change, right =purchases_of_equities,
                  on = ['티커', '종목명'], how = 'outer') # outer
print(merge3.shape)
merge3.head()
````

### 결측치 처리
- 제거 : dropna
- 대체 : fillna

````python
import numpy as np
concat_df2.loc['2023-01-05', 'LG전자':'LG화학'] = np.NaN
concat_df2.loc['2023-01-01', 'LG화학'] = np.NaN
concat_df2
````

````python
concat_df2.dropna() # axis = 0이 생략
````

````python
concat_df2.dropna(axis = 1) # axis = 0이 생략
````

````python
concat_df2.fillna(0)
````

````python
concat_df2.fillna(method = 'ffill') # 이전에 등장한 값으로 NaN을 채웁니다.
````

````python
concat_df2.fillna(method = 'bfill') # 이후에 등장한 값으로 NaN을 채웁니다.
````
