# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(100, 300, 10)
y = np.random.randn(len(x))

plt.plot(x, y)
```

```python
# figure, axes, axis
plt.figure(figsize = (15,5))
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
```

```python
plt.plot(x, y, color = 'r', linestyle = '--', marker = 'x')
```

## 여러개의 그래프 그리기

### 방법1

```python
# figure : 도화지, axes : 공간 , axis : 축
fig = plt.figure(figsize = (16,10))

axes1 = fig.add_subplot(2,2,1) # 2 x 2 사이즈에서 첫번째 위치
axes1.plot(x, y, color = 'r', linestyle = '--', marker = 's')

axes2 = fig.add_subplot(2,2,2) # 2 x 2 사이즈에서 두번째 위치
axes2.hist(y)

axes3 = fig.add_subplot(2,2,3) # 2 x 2 사이즈에서 세번째 위치
axes3.scatter(x, y)
```

### 방법 2

```python
fig, axes = plt.subplots(2,2 , figsize = (15,10))
axes[0,0].plot(x, y)
axes[1,0].hist(y)
axes[0,1].scatter(x, y)
axes[1,1].plot(x, y, 'g--o')
```

## 하나의 figure안에 두개의 그래프 그리기

```python
# figure : 도화지, axes : 공간 , axis : 축
x2 = np.arange(100, 300, 10)
y2 = np.random.randn(len(x))

fig, axes = plt.subplots(2,2 , figsize = (15,10))
axes[0,0].plot(x, y, label = 'x1')
axes[0,0].plot(x2, y2, label = 'x2')
axes[0,0].legend() # 범례표시

axes[1,0].hist(y)
axes[0,1].scatter(x, y)
axes[1,1].plot(x, y, 'g--o')
```

## 여러 그래프의 label, title 다루기

```python
# figure : 도화지, axes : 공간 , axis : 축
x2 = np.arange(100, 300, 10)
y2 = np.random.randn(len(x))

fig, axes = plt.subplots(2,2 , figsize = (15,10))
axes[0,0].plot(x, y, label = 'x1')
axes[0,0].plot(x2, y2, label = 'x2')
axes[0,0].legend() # 범례표시
axes[0,0].set_title('axes1')

axes[1,0].hist(y)
axes[1,0].set_title('axes2')

axes[0,1].scatter(x, y)
axes[0,1].set_title('axes3')

axes[1,1].plot(x, y, 'g--o')
axes[1,1].set_title('axes4')

fig.suptitle('Figure Title')
```

## 그려보기

`mpg.csv` 파일은 다양한 자동차에 대한 데이터를 포함하고 있으며, 각 행은 특정 자동차 모델을 나타냅니다. 데이터셋의 칼럼에 대한 설명은 다음과 같습니다:

1. **manufacturer**: 자동차 제조사의 이름입니다 (예: Audi).
2. **model**: 자동차의 모델입니다 (예: A4).
3. **displ**: 엔진 배기량을 리터 단위로 나타낸 것입니다 (예: 1.8, 2.0). 일반적으로 값이 크면 엔진이 더 강력합니다.
4. **year**: 자동차 모델의 제조 연도입니다 (예: 1999, 2008).
5. **cyl**: 자동차 엔진의 실린더 개수입니다 (예: 4, 6).
6. **trans**: 변속기의 종류를 나타내며, 기어 수와 변속 스타일을 포함합니다 (예: auto(l5), manual(m5)). "Auto"는 자동 변속기를, "manual"은 수동 변속기를 의미합니다.
7. **drv**: 구동 방식을 나타내며, 'f'는 전륜 구동, 'r'은 후륜 구동, '4'는 사륜 구동을 의미합니다.
8. **cty**: 도시에서의 연비를 마일 당 갤런(mpg) 단위로 나타낸 것입니다.
9. **hwy**: 고속도로에서의 연비를 마일 당 갤런(mpg) 단위로 나타낸 것입니다.
10. **fl**: 연료의 종류입니다 (예: 'p'는 프리미엄).
11. **category**: 자동차의 카테고리입니다 (예: 컴팩트, SUV).

```python
import pandas as pd
mpg = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/mpg.csv')
mpg.head()
```

```python
# 자동차 제조사별 엔진 배기량과 고속도로 연비 관계 분석
plt.scatter(mpg['displ'], mpg['hwy'])
plt.xlabel('displ')
plt.ylabel('hwy')
```

```python
# 도시 연비의 분포 분석
plt.hist(mpg['cty'])
plt.xlabel('cty')
plt.ylabel('count')
```

# seaborn

```python
# 자동차 제조사별 엔진 배기량과 고속도로 연비 관계 분석
import seaborn as sns
sns.scatterplot(data=mpg, x="displ", y="hwy")
```

```python
sns.scatterplot(data=mpg, x="displ", y="hwy", hue = 'drv')
```

```python
# 구동방식 별로 고속도로 연비를 비교
sns.barplot(data = mpg, x = 'drv', y = 'hwy')
```

```python
# 연료종류에 따른 배기량의 분포
sns.violinplot(data=mpg, x="fl", y="displ")
```

```python
# 구동방식에 따른 고속도로 연비
sns.boxplot(data=mpg, x="drv", y="hwy")
```

# plotly

```python
# 엔진 배기량과 고속도로 연비 관계 분석
import plotly.express as px
fig = px.scatter(mpg, x="displ", y="hwy")
fig.show()
```

```python
import plotly.express as px

fig = px.scatter(mpg, x="displ", y="hwy", color="drv", symbol="drv")
fig.show()
```

```python
import plotly.express as px

fig = px.scatter(mpg, x="displ", y="hwy", color="drv", symbol="drv",
                 labels = dict(displ = 'displ(x)', hwy = 'hwy(y)'))
fig.show()
```

```python
# 구동방식에 따른 고속도로 연비

fig = px.box(mpg, y="hwy", x = 'drv')
fig.show()
```
